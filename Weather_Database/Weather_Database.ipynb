```python
import random
import time
import numpy as np
import matplotlib.pyplot as plt
import requests
from config import weather_api_key
from citipy import citipy
```

```python
# Create random lat/lng combinations
lats = np.random.uniform(low = -90.000, high = 90.000, size = 2000)
lngs = np.random.uniform(low = -180.000, high = 180.000, size = 2000)
coordinates = zip(lats, lngs)
```

```python
cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

# Create storage for imported data
city_data = []
```

```python
# Log data collection
print("Data Collection Start")
print("------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop through cities
for i, j in enumerate(cities):
    
    # Log in sets of 50
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(1)
        
    # Create endpoint URL with each city
    city_url = url + "&q=" + j.replace(" ","+")
    
    # Output logged city
    print(f"Processing Record {record_count} of Set {set_count} | {j}")
    
    # Increment record count
    record_count += 1
    
    # Run an API request for each city
    try:
        # Parse the JSON
        city_weather = requests.get(city_url).json()
        
        # Parse required data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_tmp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_descript = city_weather["weather"][0]["description"]
        city_country = city_weather["sys"]["country"]
        
        # Append the city information into city_data list
        city_data.append({"City": j.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_tmp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_descript})
    
    except:
        print("City not found. Skipping...")
        pass

# Loading complete
print("----------------------")
print("Data Collection End")
print("----------------------")
```

```python
import pandas as pd
```

```python
city_data_df = pd.DataFrame(city_data)
new_column_order = ["City", "Country", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]
city_data_df = city_data_df[new_column_order]
city_data_df.head(10)
```

```python
output_file = "WeatherPy_Database.csv"
city_data_df.to_csv(output_file)
```
